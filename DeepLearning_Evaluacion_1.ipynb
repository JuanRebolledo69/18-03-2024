```python
import pandas as pd

df = pd.read_csv("IMDB Dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['sentiment'].value_counts()
```

Luego de Cargar el Dataset, se indica que contiene 50.000 reseñas, 25.000 Positivas y 25.000 Negativas, no hay Nulos.

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=5000)

X = vectorizer.fit_transform(df['review']).toarray()
```

```python
X.shape
```

Usamos "CountVectorizer" para transformar las palabras a representaciones numericas y que de esa forma el modelo pueda entenderlas, limitado a una cantidad maxima de 5.000 palabras.

```python
y = df['sentiment'].map({'positive':1, 'negative':0})
```

```python
y.head()
```

Ahora transformamos las reseñas a numeros para la comprension del Modelo. Las Positivas como "1" y las Negativas como "0".

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

Se dividio el Dataset en dos partes y se asigno el 80% de los datos para entrenamiento y el 20% restante para pruebas y evaluacion. De esa forma comprobamos si el Modelo realmente aprendio o solo memorizo.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(16, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

Se contruye una red neuronal con Keras, con una arquitectura que consiste en una capa de entrada, conectada a una capa oculta de 16 neuronas con activacion ReLU, que continua a una segunda capa oculta de 8 neuronas y finalmente se utiliza la capa de salida con una neurona y Sigmoide para realizar la clasificacion binaria de Positiva o Negativo.

```python
model.summary()
```

80.161 Parametros entrenables.

<H1> Entrenamiento 1: <H1>

3 Epocas / Batch 32

```python
model.fit(X_train, y_train, epochs=3, batch_size=32)
```

```python
model.evaluate(X_test, y_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)

print(classification_report(y_test, y_pred))
```

Se evaluo al Modelo con el conjunto de datos apartados para Pruebas con 3 Epocas y se obtuvo una presicion (accuracy) cercana al 88% de acierto y un valor de perdida (loss) de 0.30, un valor relativamente bajo. Este resultado indica que el Modelo es capaz de clasificar correctamente la mayoria de las reseñas.

<H1> Entrenamiento 2: <H1>

5 Epocas / Batch 32

```python
model.fit(X_train, y_train, epochs=5, batch_size=32)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)

print(classification_report(y_test, y_pred))
```

Luego de entrenar por segunda vez el modelo, esta vez con 5 Epocas, la presicion bajo a 87% y el valor de perdida (loss) subio a 0.53.

<H1> Entrenamiento 3: <H1>

3 Epocas / Batch 64

```python
model.fit(X_train, y_train, epochs=3, batch_size=64)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)

print(classification_report(y_test, y_pred))
```

Esta vez, nuevamente entrenamos al Modelo con 3 Epocas, pero cambiamos el tamaño del Batch a 64, dando como resultado una precision del 87% y un valor de perdida (loss) de 0.80.

<h1> Comparacion de los 3 Entrenamientos <h1>

Luego de Entrenar al Modelo tres veces variando la cantidad de Epocas y el tamaño del Batch, podemos ver que la version con 3 Epocas y Batch de 32 dio un mejor resultado con un 88% de presicion y un valor de perdida de 0.30, mientras que al entrenarlo con mas Epocas y mas tamaño de Batch, su presicion dsiminuyo y su valor de perdida (loss) aumento.

Esto sugiere que entrenar al modelo mas tiempo, no necesariamente mejora su rendimiento.
